# 3.Write a python program to access the search bar and search button on images.google.com and scrape 10images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
def image(x):
    driver=webdriver.Chrome("chromedriver.exe")
    time.sleep(5)
    driver.maximize_window()
    url='http://images.google.com/'
    #now we want to open a webpage of the given url
    driver.get(url)
    #now we take "fruits" as a user input
    search_bar=driver.find_element_by_xpath("/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input")
    search_bar.send_keys(x)
    time.sleep(2)
    #clicking search button
    search_button=driver.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div')
    search_button.click()
    time.sleep(2)
    images=driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')
    image=[]
    for i in images:
        image.append(i.get_attribute('src'))
    df=pd.DataFrame({'Image Link':image[0:10]})
    return(df)
    driver.close()
```

# 5.Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on googlemaps

```python
#Define function to find coordinates (latitude, longitude) of a city searched on googlemaps
def find_cordinate(location):
    # connecting to the webdriver
    driver=webdriver.Chrome("chromedriver.exe")
    driver.maximize_window()

    url='https://www.google.com/maps/'
    #oening the webpage
    driver.get(url)
    #finding the search bar
    search_bar= driver.find_element_by_xpath('/html/body/div[3]/div[9]/div[3]/div[1]/div[1]/div[1]/div[2]/form/div/div[3]/div/input[1]')
    #passing the location to the search bar
    search_bar.send_keys(location)

    # Click search button
    search_button=driver.find_element_by_xpath('/html/body/div[3]/div[9]/div[3]/div[1]/div[1]/div[1]/div[2]/div[1]/button')
    search_button.click()
    time.sleep(5)
    
    url=driver.current_url
    #finding Latitude and longitude from the current url
    p=url.split('@')[1].split(',')[0:2]
    #printing the Latitude and Longitude
    print(" Latitude:",p[0],"\n","Longitude:",p[1])
    #closing the driver
    driver.close()
    return
```

```python
#finding coordinates (latitude, longitude) of "Kolkata"
find_cordinate('Kolkata')
```

```python
#finding coordinates (latitude, longitude) of another city "Bangalore"
find_cordinate('Bangalore')
```

You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description

```python
#importing required classes
from selenium import webdriver
```

```python
# connecting to the webdriver
driver=webdriver.Chrome("chromedriver.exe")
```

```python
# getting the web page of mentioned url
url = "https://www.hostelworld.com/"
driver.get(url)
time.sleep(3)
```

```python
# finding the search bar
search_bar = driver.find_element_by_xpath("/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/div/input")

# passing London in search bar
search_bar.send_keys("London")
```

```python
# select London
London = driver.find_element_by_xpath("//ul[@id='predicted-search-results']//li[2]")
#clicking on button
London.click()

# do click on Let's Go button
search_btn = driver.find_element_by_id('search-button')
search_btn.click()
```

```python
# Make empty lists
Hostel_Name = []
Distance = []
overall_review = []
total_reviews = []
facilities = []
price = []
Rating = []
property_description = []
```

```python
while(True):
    
    # scrap Hostel name
    names = driver.find_elements_by_xpath('//h2[@class="title title-6"]')
    for name in names:
        Hostel_Name.append(name.text)
    time.sleep(2)
        
    # scrap Distance from city centre
    dis = driver.find_elements_by_xpath('//span[@class="description"]')
    for d in dis:
        Distance.append(d.text)
    time.sleep(2)
        
    #scrap Overall Review    
    review = driver.find_elements_by_xpath('//div[@class="keyword"]//span')
    for r in review:
        overall_review.append(r.text)
    time.sleep(2)
    
    #scrap Total No of reviews     
    t_review = driver.find_elements_by_xpath('//div[@class="reviews"]')
    for tr in t_review:
        total_reviews.append(tr.text)
    time.sleep(2)
    
    #scrap facilities
    service = driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
    for s in service:
        facilities.append(s.text)
    time.sleep(2)
    
    #scrap Prices    
    prices = driver.find_elements_by_xpath('//div[@class="price-col"]')
    for p in prices:
        price.append(p.text)
    time.sleep(2)    
        
    try:
        next_button = driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
        
time.sleep(4)        
        
# Separate  Privates_From price  and  Dorms_From price
Privates_From = []
for i in range(0,len(price),2):
    Privates_From.append(price[i])
time.sleep(2)

Dorms_From = []
for i in range(1,len(price),2):
    Dorms_From.append(price[i])
```

```python
# scrape Hostels URL
hostel_url = []

while(True):
    urls = driver.find_elements_by_xpath('//h2[@class="title title-6"]//a')
    for url in urls:
        hostel_url.append(url.get_attribute("href"))
    time.sleep(2)    
        
    try:
        next_button = driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break


        
Rate = []
for page in hostel_url:
    driver.get(page)
    
    # Rating
    try:
        ratings = driver.find_element_by_xpath('/html/body/div[1]/div/div/div[1]/section/div[6]/div/div[1]/div[1]/div[1]')
        Rate.append(ratings.text)
    except NoSuchElementException:
        Rate.append("No Rating")  
    time.sleep(2)
    
    
    # Property Description
    try:
        pd = driver.find_element_by_xpath('/html/body/div[1]/div/div/div[1]/section/div[6]/div/div[2]/div[2]/div/div[2]')
        property_description.append(pd.text)
    except NoSuchElementException:
        property_description.append("No Description")  

    
time.sleep(2)        
# remove extra data from Rating     
all_text = []
for i in Rate:
    all_text.append(i.split())
time.sleep(2)

for i in all_text:
    Rating.append(i[0]) 
```
